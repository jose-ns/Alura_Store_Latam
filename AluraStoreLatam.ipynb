### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto
ingreso_total_1 = tienda['Precio'].sum()
print(f'Ingresos tienda 1: {ingreso_total_1}')

ingreso_total_2 = tienda2['Precio'].sum()
print(f'Ingresos tienda 2: {ingreso_total_2}')

ingreso_total_3 = tienda3['Precio'].sum()
print(f'Ingresos tienda 3: {ingreso_total_3}')

ingreso_total_4 = tienda4['Precio'].sum()
print(f'Ingresos tienda 4: {ingreso_total_4}')

ventas_totales = [ingreso_total_1, ingreso_total_2, ingreso_total_3, ingreso_total_4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

#La tienda 1 está facturando más, en cambio, la tienda 4 menos
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(ventas_totales, labels=nombres_tiendas, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Ventas por Tienda')
plt.axis('equal')
plt.show()
```

# 2. Ventas por categoría

```python
#agrupar los datos por categoría y contar el número de ventas de cada tipo, orden descendente
productos_por_categoria_1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
print('Tienda 1')
print('Categorías más populares:')
for producto, cantidad in productos_por_categoria_1.items():
    print(f'- {producto}: {cantidad}')
print()

productos_por_categoria_2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
print('Tienda 2')
print('Categorías más populares:')
for producto, cantidad in productos_por_categoria_2.items():
    print(f'- {producto}: {cantidad}')
print()

productos_por_categoria_3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
print('Tienda 3')
print('Categorías más populares:')
for producto, cantidad in productos_por_categoria_3.items():
    print(f'- {producto}: {cantidad}')
print()

productos_por_categoria_4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print('Tienda 4')
print('Categorías más populares:')
for producto, cantidad in productos_por_categoria_4.items():
    print(f'- {producto}: {cantidad}')
print()
```

```python
data_categorias = pd.DataFrame({
    'Tienda 1': productos_por_categoria_1,
    'Tienda 2': productos_por_categoria_2,
    'Tienda 3': productos_por_categoria_3,
    'Tienda 4': productos_por_categoria_4,
}).fillna(0)

data_categorias.plot(kind='bar', figsize=(10, 6), title='Comparación de Categorías por Tienda')
plt.ylabel('Cantidad de Productos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
#calcular las calificaciones promedio de los clientes para cada tienda

promedio_tienda_1 = (tienda['Calificación'].sum())/(len(tienda['Calificación']))
promedio_1 = round(promedio_tienda_1, 2)
print(f'La calificación promedio de la tienda 1 es: {promedio_1}')

promedio_tienda_2 = (tienda2['Calificación'].sum())/(len(tienda2['Calificación']))
promedio_2 = round(promedio_tienda_2, 2)
print(f'La calificación promedio de la tienda 2 es: {promedio_2}')

promedio_tienda_3 = (tienda3['Calificación'].sum())/(len(tienda3['Calificación']))
promedio_3 = round(promedio_tienda_3, 2)
print(f'La calificación promedio de la tienda 3 es: {promedio_3}')

promedio_tienda_4 = (tienda4['Calificación'].sum())/(len(tienda4['Calificación']))
promedio_4 = round(promedio_tienda_4, 2)
print(f'La calificación promedio de la tienda 4 es: {promedio_4}')

#La tienda 2,3 y 4 tienen la mejor calificación promedio, la tienda 1 la menor
```

# 4. Productos más y menos vendidos

```python
productos_vendidos_1 = tienda.groupby('Producto').size().sort_values(ascending=False)
mas_ventas_1 = productos_vendidos_1.head(3)
menos_ventas_1 = productos_vendidos_1.tail(3)
print('Tienda 1')
print('Más vendidos:')
for producto, cantidad in mas_ventas_1.items():
    print(f'- {producto}: {cantidad}')
print('Menos vendidos:')
for producto, cantidad in menos_ventas_1.items():
    print(f'- {producto}: {cantidad}')
print()

productos_vendidos_2 = tienda2.groupby('Producto').size().sort_values(ascending=False)
mas_ventas_2 = productos_vendidos_2.head(3)
menos_ventas_2 = productos_vendidos_2.tail(3)
print('Tienda 2')
print('Más vendidos:')
for producto, cantidad in mas_ventas_2.items():
    print(f'- {producto}: {cantidad}')
print('Menos vendidos:')
for producto, cantidad in menos_ventas_2.items():
    print(f'- {producto}: {cantidad}')
print()

productos_vendidos_3 = tienda3.groupby('Producto').size().sort_values(ascending=False)
mas_ventas_3 = productos_vendidos_3.head(3)
menos_ventas_3 = productos_vendidos_3.tail(3)
print('Tienda 3')
print('Más vendidos:')
for producto, cantidad in mas_ventas_3.items():
    print(f'- {producto}: {cantidad}')
print('Menos vendidos:')
for producto, cantidad in menos_ventas_3.items():
    print(f'- {producto}: {cantidad}')
print()


productos_vendidos_4 = tienda4.groupby('Producto').size().sort_values(ascending=False)
mas_ventas_4 = productos_vendidos_4.head(3)
menos_ventas_4 = productos_vendidos_4.tail(3)
print('Tienda 4')
print('Más vendidos:')
for producto, cantidad in mas_ventas_4.items():
    print(f'- {producto}: {cantidad}')
print('Menos vendidos:')
for producto, cantidad in menos_ventas_4.items():
    print(f'- {producto}: {cantidad}')
print()
```

```python
ventas_destacadas_1 = pd.concat([mas_ventas_1, menos_ventas_1])

ventas_destacadas_1.plot(kind='bar', color=['green']*3 + ['red']*3, title='Tienda 1 - Más y Menos Vendidos')
plt.ylabel('Cantidad vendida')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

ventas_destacadas_2 = pd.concat([mas_ventas_2, menos_ventas_2])

ventas_destacadas_2.plot(kind='bar', color=['green']*3 + ['red']*3, title='Tienda 2 - Más y Menos Vendidos')
plt.ylabel('Cantidad vendida')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

ventas_destacadas_3 = pd.concat([mas_ventas_3, menos_ventas_3])

ventas_destacadas_3.plot(kind='bar', color=['green']*3 + ['red']*3, title='Tienda 3 - Más y Menos Vendidos')
plt.ylabel('Cantidad vendida')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

ventas_destacadas_4 = pd.concat([mas_ventas_4, menos_ventas_4])

ventas_destacadas_4.plot(kind='bar', color=['green']*3 + ['red']*3, title='Tienda 4 - Más y Menos Vendidos')
plt.ylabel('Cantidad vendida')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
#calcular el costo de envío promedio para cada tienda
print('El costo de envío promedio en tienda 1 es:', round(tienda['Costo de envío'].mean()))
print('El costo de envío promedio en tienda 2 es:', round(tienda2['Costo de envío'].mean()))
print('El costo de envío promedio en tienda 3 es:', round(tienda3['Costo de envío'].mean()))
print('El costo de envío promedio en tienda 4 es:', round(tienda4['Costo de envío'].mean()))

#La tienda 1 tiene un mayor costo pormedio de envíó, la tienda 4 el menor costo
```

```python
promedios_envio = [
    round(tienda['Costo de envío'].mean(), 2),
    round(tienda2['Costo de envío'].mean(), 2),
    round(tienda3['Costo de envío'].mean(), 2),
    round(tienda4['Costo de envío'].mean(), 2)
]

nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(9, 6))
barras = plt.barh(nombres_tiendas, promedios_envio, color='lightgreen')

for barra in barras:
    ancho = barra.get_width()
    plt.text(ancho + 0.5, barra.get_y() + barra.get_height()/2,
             f"${ancho}", va='center')

plt.title('Costo de envío promedio por tienda')
plt.ylabel('Tienda')
plt.xlabel('Costo promedio de envío ($)')
plt.xlim(0, max(promedios_envio) + 10)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
